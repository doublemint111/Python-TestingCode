```python
Test02=Test.loc[Test.year==2007]
w=len(Test02)
for i in range(w):
    pop=pop+Test02.iloc[i,4]
pop
```

#### Workshop 1-7: 計算 2007 年全球的平均壽命、平均財富

```python
import scipy
Test03=Test.loc[(Test.year==2007)]
y1=Test03.lifeExp
y2=Test.gdpPercap
print('平均壽命:')
print(scipy.mean(y1))
print('平均財富:')
print(scipy.mean(y2))
```

```python
name = ["蒙其·D·魯夫", "羅羅亞·索隆", "娜美", "騙人布", "賓什莫克·香吉士", "多尼多尼·喬巴", "妮可·羅賓", "佛朗基", "布魯克"]
occupation = ["船長", "劍士", "航海士", "狙擊手", "廚師", "醫生", "考古學家", "船匠", "音樂家"]

# 建立 dict
straw_hat_dict = {"name": name,
                  "occupation": occupation
}

# 使用 straw_hat_dict 建立一個 data frame
straw_hat_df = pd.DataFrame(straw_hat_dict)
straw_hat_df
```

```python
# 建立第二個 data frame
name = ["蒙其·D·魯夫", "多尼多尼·喬巴", "妮可·羅賓", "布魯克", "Michael"]
devil_fruit = ["橡膠果實", "人人果實", "花花果實", "黃泉果實", "芭樂果實"]

# 建立 dict
devil_fruit_dict = {"name": name,
                    "devil_fruit": devil_fruit
}

# 建立第二個 data frame devil_fruit_df
devil_fruit_df = pd.DataFrame(devil_fruit_dict)
devil_fruit_df
```

```python
# 合併兩個 dataframe，類似 DB 的 inner join
straw_hat_merged = pd.merge(straw_hat_df, devil_fruit_df)
straw_hat_merged
```

```python
# left outer join 合併兩個 dataframe 
straw_hat_left_merged = pd.merge(straw_hat_df, devil_fruit_df, how='left')
straw_hat_left_merged
```

```python
# right outer join 合併兩個 dataframe 
straw_hat_left_merged = pd.merge(straw_hat_df, devil_fruit_df, how='right')
straw_hat_left_merged
```

```python
# full outer join 合併兩個 dataframe 
straw_hat_left_merged = pd.merge(straw_hat_df, devil_fruit_df, how='outer')
straw_hat_left_merged
```
