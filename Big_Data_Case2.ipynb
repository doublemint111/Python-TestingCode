## 案例1-2: 世界各國幾個年度的人口數、平均壽命和平均財富的範例
- 資料集下載點: https://storage.googleapis.com/learn_pd_like_tidyverse/gapminder.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Workshop 1-11: (繪製直線圖) 將臺灣資料篩選出來並繪製從 1952 年至 2007 年人口變化的直線圖

```python
df = pd.read_csv('https://storage.googleapis.com/learn_pd_like_tidyverse/gapminder.csv')
```

```python
df2 = df[df['country'] == 'Taiwan']
df2[['year', 'pop']].plot(kind = 'line', x = 'year', y = 'pop', title = 'Pop vs. Year in Taiwan', legend = False)
plt.show()
```

```python
df['year'].unique()
```

#### Workshop 1-12: (繪製直線圖) 將中國、日本、南韓與臺灣資料篩選出來，並繪製從 1952 年至 2007 年平均壽命變化的直線圖

```python
gapminder_northasia = df.loc[df['country'].isin(['China', 'Japan', 'Korea, Rep.', 'Taiwan'])]
gapminder_northasia_pivot = gapminder_northasia.pivot_table(values = 'lifeExp', columns = 'country', index = 'year')
gapminder_northasia_pivot.plot(title = 'Life Expectancies in North Asia')
plt.show()
```

#### Workshop 1-13: (繪製長條圖) 繪製 2007 年全球各洲的人口總數

```python
gapminder_2007 = df[df['year'] == 2007]
gapminder_2007[['pop', 'gdpPercap', 'lifeExp']].hist(bins = 15)
plt.show()
```

```python
summarized_df = df[df['year'] == 2007].groupby(by = 'continent')['pop'].sum()
summarized_df.plot(kind = 'bar', rot = 0)
plt.show()
```

#### Workshop 1-14: (繪製長條圖) 將中國、日本、南韓與臺灣資料篩選出來，並繪製從 1952 年至 2007 年平均壽命變化的長條圖

```python
df3=df.loc[df['country'].isin(['China', 'Japan', 'Korea, Rep.', 'Taiwan'])]
df4=df3.pivot_table(values = 'lifeExp', columns = 'country', index = 'year')

df4.plot(kind = 'bar', rot = 0,title = 'Life Expectancies in North Asia')
plt.show()
```

#### Workshop 1-15: (繪製垂直的長條圖) 繪製各洲的平均財富 & 平均壽命 (製作 1 x 2 的 subplots )

```python
summarized_df = df[df['year'] == 2007].groupby(by = 'continent')[['lifeExp', 'gdpPercap']].mean()
summarized_df.plot(kind = 'barh', subplots = True, layout = (1, 2), sharex = False, sharey = True, legend = False)
plt.show()
```
