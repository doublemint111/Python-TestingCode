```python
#請建立三個一維的整數 numpy.ndarray 分別是 'mathematics', 'english', 以及 'chinese'，範圍介於 0 - 100，各有 10 筆資料。
import random
mathematics=[]
english=[]
chinese=[]
for i in range(10):
    mathematics.append(random.randint(1, 100)) 
    english.append(random.randint(1, 100)) 
    chinese.append(random.randint(1, 100)) 
    
print(mathematics)
print(english)
print(chinese)
```

```python
#請建立一個 從 'a1' 到 'a10' 的串列，名稱為 'idx'。
idx = np.arange(1, 11)
idx = ['a' + str(sid) for sid in idx]
```

```python
#請建立一個名為 score 的資料框，score 的index是 idx, 三個欄位分別是'mathematics', 'english', 以及 'chinese'。
score = pd.DataFrame({'mathematics':mathematics, 'english':english, 'chinese':chinese}, index=idx)
```

```python
#請計算每位同學的平均成績，並新增為 'average' 欄位。
score['average'] = score.mean(axis=1) # axis=1 表示以列為基準，求取每列的平均值
```

```python
#請依照每位同學的平均成績顯示每個人的名次，並新增為 'rank' 欄位 (請自行研究 rank() 的用法)。
#rank 相關用法可以查看 -- https://blog.csdn.net/starter_____/article/details/79183595
rank = score.average.rank(ascending=False, method='max')#降序排列 ascending=False, method='max'
score['rank'] = rank
score['rank']= score['rank'].astype('int')
score.sort_values(['rank'],ascending='flase')
```
