```python
#Workshop 11 綜和演練題
#請建立一個 score DataFrame，有 20 列 5 門科目 (20 x 5 的矩陣)，請使用所學過的隨機方法填入成績
import random
S1=[]
S2=[]
S3=[]
S4=[]
S5=[]
def randomScore(S):
    S.append(random.randint(30, 100))
    return S

for i in range(20):
    S1=randomScore(S1)
    S2=randomScore(S2)
    S3=randomScore(S3)
    S4=randomScore(S4)
    S5=randomScore(S5)

ScoreDict={'物理':S1,'化學':S2,'數學':S3,'英文':S4,'藝術':S5}
df=pd.DataFrame(ScoreDict)

idx = {}
for i in range(len(df)):
    idx[i]=(str(i + 1))
    
df=df.rename(index=idx)

df.head()
```

```python
#請計算每一位同學的平均成績，並新增每位學生的平均成績到總平均欄位，並請排序總成績，新增排序結果到總成績排序欄位
DFC=df.columns
TotalAva=[]
Sum=0
for i in range(len(df)):
    Sum=0
    Ava=0
    for G in DFC:
        Sum=Sum+(df[G].iloc[i])
        Ava=Sum/5
        Ava='%.2f' % Ava
    TotalAva.append(Ava)

ScoreDict={'物理':S1,'化學':S2,'數學':S3,'英文':S4,'藝術':S5,'總平均':TotalAva}
df=pd.DataFrame(ScoreDict)
rank = df.總平均.rank(ascending=False, method='max')
df['總成績排行'] = rank
df['總成績排行']= df['總成績排行'].astype('int')
df.sort_values(['總成績排行'],ascending='flase')
```

```python
#也請將每門課的成績排序，並存成例如物理排序、化學排序等名稱，依序插入每門課的右側
TotalRank = df.總平均.rank(ascending=False, method='max')
S1Rank = df.物理.rank(ascending=False, method='max')
S2Rank = df.化學.rank(ascending=False, method='max')
S3Rank = df.數學.rank(ascending=False, method='max')
S4Rank = df.英文.rank(ascending=False, method='max')
S5Rank = df.藝術.rank(ascending=False, method='max')
ScoreDict={'物理':S1,'物理排序':S1Rank,'化學':S2,'化學排序':S2Rank,'數學':S3,'數學排序':S3Rank,'英文':S4,'英文排序':S4Rank,'藝術':S5,'藝術排序':S5Rank,'總平均':TotalAva}
df=pd.DataFrame(ScoreDict)
df['總成績排行'] = TotalRank
df['總成績排行']= df['總成績排行'].astype('int')
df['物理排序']= df['物理排序'].astype('int')
df['化學排序']= df['化學排序'].astype('int')
df['數學排序']= df['數學排序'].astype('int')
df['英文排序']= df['英文排序'].astype('int')
df['藝術排序']= df['藝術排序'].astype('int')

df.sort_values(['總成績排行'],ascending='flase')
```

```python
#請將 score 資料框寫入檔案，分別存成 score.csv 以及 score.xlsx。
import csv
#創建csv檔案
with open('Score.csv','w',newline='')as files:
    writer = csv.writer(files)
    writer.writerow(df.columns)
    
    for i in range(len(df)):
        writer.writerow(df.iloc[i]) 
```

```python
#請再建立一個新的20 列 5 門科目 (20 x 5 的矩陣) 的 score2 DataFrame。並在 20 x 5 的成績隨機插入 3 個 0 分的成績。
S1=[]
S2=[]
S3=[]
S4=[]
S5=[]
def randomScore(S):
    S.append(random.randint(30, 100))
    return S

for i in range(20):
    S1=randomScore(S1)
    S2=randomScore(S2)
    S3=randomScore(S3)
    S4=randomScore(S4)
    S5=randomScore(S5)

ScoreDict2={'物理':S1,'化學':S2,'數學':S3,'英文':S4,'藝術':S5}
df2=pd.DataFrame(ScoreDict2)

for i in range(3):
    x=random.randint(0,19)
    y=random.randint(0,4)
    df2.iloc[x,y]=0
    
idx = {}
for i in range(len(df2)):
    idx[i]=(str(i + 1))
    
df2=df2.rename(index=idx)

#將修改完成的資料存起來
S1=df2.iloc[:20,0]
S2=df2.iloc[:20,1]
S3=df2.iloc[:20,2]
S4=df2.iloc[:20,3]
S5=df2.iloc[:20,4]


df2.head(20)
```

```python
#請將計算總平均、總成績排序，以及物理排序、化學排序等各科目的排序寫成一個名為 calcScore 的函數。
#將 score2 當成參數傳遞到 calcScore，計算完畢後回傳結果。

def calcScore(score2):
    DFC=score2.columns
    TotalAva=[]
    #傳入資料表作為參數依據(函數內稱傳入資料表為score2)分析判斷加總值為多少並記錄
    for i in range(len(score2)):
        Sum=0
        Ava=0
        for G in DFC:
            Sum=Sum+(score2[G].iloc[i])
            Ava=Sum/5
            Ava='%.2f' % Ava
        TotalAva.append(Ava)
    
    ScoreDict2={'物理':S1,'化學':S2,'數學':S3,'英文':S4,'藝術':S5,'總平均':TotalAva}
    score2=pd.DataFrame(ScoreDict2)
    
    #分析資料表製作出各項Rank排行並記錄
    TotalRank = score2.總平均.rank(ascending=False, method='max')
    S1Rank = score2.物理.rank(ascending=False, method='max')
    S2Rank = score2.化學.rank(ascending=False, method='max')
    S3Rank = score2.數學.rank(ascending=False, method='max')
    S4Rank = score2.英文.rank(ascending=False, method='max')
    S5Rank = score2.藝術.rank(ascending=False, method='max')
    ScoreDict2={'物理':S1,'物理排序':S1Rank,'化學':S2,'化學排序':S2Rank,'數學':S3,'數學排序':S3Rank,'英文':S4,'英文排序':S4Rank,'藝術':S5,'藝術排序':S5Rank,'總平均':TotalAva}
    score2=pd.DataFrame(ScoreDict2)
    score2['總成績排行'] = TotalRank
    score2['總成績排行']= score2['總成績排行'].astype('int')
    score2['物理排序']= score2['物理排序'].astype('int')
    score2['化學排序']= score2['化學排序'].astype('int')
    score2['數學排序']= score2['數學排序'].astype('int')
    score2['英文排序']= score2['英文排序'].astype('int')
    score2['藝術排序']= score2['藝術排序'].astype('int')
    

    #把各項紀錄資料重新寫入資料表並將最終的資料表傳回給呼叫者
    return (score2.sort_values(['總成績排行'],ascending='flase'))

    

    
```
