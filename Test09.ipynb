```python
# 再重新產製一 個 5 x 3 的 DataFrame
d = np.random.randint(1, 100, (5, 3))
df = pd.DataFrame(d, columns=['Aa', 'Bb', 'Cc'])

# 在 df 隨機設定 3 個 NaN
for i in range(3):
    x = random.randint(0, 4)
    y = random.randint(0, 2)
    df.iloc[x,y] = np.nan

# 再新增一列全是 NaN 的資料列
df = df.append({'Aa':np.nan, 'Bb': np.nan, 'Cc': np.nan}, ignore_index=True)
df
```

```python
# 計算每一列的平均值
m = df.mean(axis=1)
m
```

```python
# 將每列的平均值填入每列的 NaN 欄位 (想不到更好用的辦法...只好用 loop... :( )
for i, item in enumerate(m):
    if(df.loc[i].isnull().any()):
        df.loc[i].fillna(item, inplace=True)
df
```

```python
#Workshop 13
import pandas as pd
import numpy as np
Air = pd.read_csv('Air.csv')#Air.csv
Air.head()
```

```python
#請找出任何有 NaN 的資料列。
Air.isnull().sum()
```

```python
#請取出臺南市的所有資料列
AirTainan=Air.loc[Air.County=='臺南市']
AirTainan
```

```python
#請將臺南市每一種汙染源的平均值填入其 NaN 欄位
for i in range(len(AirTainan)):
    for G in range(6):
        x=2+G
        AirTainan.iloc[i,x]=np.nan
AirTainan
```

```python
#請取出高雄市的所有資料列，並請將有 NaN 的欄位以前一筆的資料填入 (method='ffill')
AirKaohsiung=Air.loc[Air.County=='高雄市']
AirKaohsiung=AirKaohsiung.fillna(method='ffill')
AirKaohsiung
```
