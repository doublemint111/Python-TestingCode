```python
# case 2: 讀入鐵達尼號的資料集，利用 map() 產生另一個欄位 Series
titanic = pd.read_csv('http://bit.ly/kaggletrain')
titanic.head()
```

```python
#Workshop 14
import pandas as pd
import numpy as np
#請使用 titanic (http://bit.ly/kaggletrain) 的 資料集。

#在 titanic.Embarked 欄位只有三個可能值，分別是 'S', 'C' 和 'Q'，們可以利用 unique 得知此欄位的三個可能結果:
#titanic.Embarked.unique()
titanic = pd.read_csv('http://bit.ly/kaggletrain')
titanic.head()
```

```python
#1.titanic.Embarked 欄位有 S, C 和 Q 三個可能值，代表鐵達尼號乘客登船的三個地點。C 是表示法國的 Cherbourg (瑟堡港); 
#Q 是表示愛爾蘭的 Queenstown (現改稱為 Cobh)；S 是表示英國的 Southampton (南安普敦)。用字母簡稱不易確認地名，
#請利用 apply() 或 map() 替換真正的名稱。
titanic['Embarked']=titanic.Embarked.apply(lambda x: 'Southampton' if x == 'S' else ('Cherbourg' if x == 'C' else 'Queenstown'))
```

```python
titanic.head()
```

```python
#2.titanic.Fare 欄位是鐵達尼號的船票價格，此欄位是浮點數型態，請將此欄位的數值取道小數點後 1 位數。
#titanic.loc[titanic.Fare]=float(titanic.loc[titanic.Fare])
#titanic['Fare']=float(titanic['Fare'])
titanic['Fare']=titanic['Fare'].apply(lambda x: '%.1f' % x if type(x)==float else x)
titanic.head()
```
